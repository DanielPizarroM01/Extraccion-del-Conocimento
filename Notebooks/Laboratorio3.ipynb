```python
import pandas as pd
```

```python
datos_instagram = {
    "influencer": ["@cancun_vibes", "@travel_mexico", "@beach_lover", "@paradise_found", "@azul_turquesa", "@explora_qroo"],
    "ciudad_tema": ["cancun", "playa_del_carmen", "tulum", "isla_mujeres", "cozumel", "bacalar"],
    "tipo_post": ["photo", "video", "reel", "carousel"],
    "fecha_registro": ["2026-01-15", "2026-02-20", "2026-03-12", "2026-04-05", "2026-05-22"]
}
```

```python
import random
```

```python
cantidad = 1000000
```

```python
# Crear el dataFrame de 1000000 de registros con la estructura solicitada
dfInstagramMillon = pd.DataFrame({
    "influencer": [random.choice(datos_instagram["influencer"]) for _ in range(cantidad)],
    "likes": [random.randint(500, 50000) for _ in range(cantidad)],
    "comentarios": [random.randint(10, 2500) for _ in range(cantidad)],
    "fecha": [random.choice(datos_instagram["fecha_registro"]) for _ in range(cantidad)],
    "tipo_post": [random.choice(datos_instagram["tipo_post"]) for _ in range(cantidad)]
})

# Verificamos la forma del DataFrame para asegurar el millón de filas
dfInstagramMillon.shape
```

```python
# 1. Crear métrica de Engagement Total
dfInstagramMillon['engagement_total'] = dfInstagramMillon['likes'] + dfInstagramMillon['comentarios']

print("--- INSIGHT 1: Engagement promedio por Tipo de Publicación ---")
print(dfInstagramMillon.groupby('tipo_post')['engagement_total'].mean().sort_values(ascending=False))

print("\n--- INSIGHT 2: Top 3 Influencers con más interacciones acumuladas ---")
print(dfInstagramMillon.groupby('influencer')['engagement_total'].sum().sort_values(ascending=False).head(3))
```
