```python
# importamos la libreria FAKER
from faker import Faker
```

```python
# Creamos una instancia faker
fake = Faker("es_MX")
```

```python
datos = []
for i in range(100):
    datos.append(
        {
            "id_usuario": i +1,
            "nombre" : fake.name(),
            "corro": fake.email(),
            "edad" : random.randint(18,65),
            "seguidores" : random.randint(50,5000),
            "plataforma" : random.choice(["tiktok","Intagram","YouTube"])
        }
    )
```

```python
dfInfluencer = pd.DataFrame(datos)
```
